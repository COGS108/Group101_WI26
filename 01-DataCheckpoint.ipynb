## Research Question

How well can season-to-season changes in professional male soccer players’ inflation-adjusted market values (as reported by Transfermarkt) from 2013–2023 be predicted using age, primary playing position (goalkeeper, defender, midfielder, forward), total league minutes played, and standardized performance metrics (goals, assists, expected goals, expected assists, key passes, defensive actions per 90)?

Among these variables, which are the strongest predictors of positive versus negative changes in market value between consecutive seasons?


## Background and Prior Work

Professional soccer operates as a global labor market in which player valuations influence transfers, wages, sponsorship visibility, and long-term club strategy. Top European leagues generate billions of dollars annually in revenue, and transfer spending alone regularly exceeds several billion euros per season.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) In this context, how players are valued has real economic consequences for clubs, agents, and players themselves. Because actual transfer fees only materialize when a transfer occurs, researchers and analysts often rely on publicly available “market value” estimates as a consistent proxy for how the market rates a player at a given moment. These values allow large-scale cross-league and cross-season statistical analysis that would not be possible using transfer fees alone. 

A widely used public source of such estimates is Transfermarkt, which publishes crowd-informed market values. Transfermarkt explicitly distinguishes its estimates from realized transfer fees and describes its valuation process as combining pricing logic with structured, moderated community discussion.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) This is crucial for our project: it implies that market value captures meaningful signals (age, contract status, performance, reputation) but may also reflect perception-based biases such as league visibility, media exposure, or hype. Thus, market value is best understood not as a “true” intrinsic worth, but as an observable outcome of a social and economic evaluation process. 

Prior academic research has investigated which factors explain player market value. A recent systematic review synthesizes evidence that performance metrics, age, positional role, contract characteristics, and contextual variables all contribute to valuation patterns.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Müller et al. (2017) developed a data-driven framework to estimate market values across multiple European leagues and demonstrated that statistical modeling can reproduce meaningful valuation patterns using performance and contextual indicators.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) Their work shows that valuation is partially predictable and structured, rather than arbitrary. 

However, much of the existing literature focuses either on specific subsets of leagues (e.g., top European competitions) or on predictive accuracy rather than interpretability. In other words, prior work often asks: Can we estimate market value? Our project instead emphasizes: Which measurable attributes are most strongly associated with market value in our dataset, and how do contextual factors compare to on-field performance? By focusing on interpretability within a regression-style framework, we aim to clarify the relative contribution of performance, demographic, and contextual predictors rather than maximizing prediction alone. 

Researchers have also documented structural effects that may influence valuation. For example, studies of relative age effects find that players born earlier in the selection year may have higher market values in youth contexts, suggesting that developmental structures shape perceived value.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) This highlights a broader concern: even when performance metrics are included, market value may embed selection effects and institutional bias. Additionally, competing “algorithmic valuation” proposals—such as those referencing CIES-style models incorporating age and contract length—demonstrate that valuation is multi-factor and contested, with no universally accepted formula.<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6) 

These prior studies motivate and delimit our contribution. Rather than proposing a new proprietary valuation model or claiming to identify a “true” player worth, our goal is to empirically examine associations between measurable player attributes (performance statistics, playing time, age, position, and league/club context) and publicly reported market value within our dataset. By situating our analysis within this broader literature, we treat market value as a socially informed economic signal and focus on interpreting patterns rather than asserting causal or normative conclusions.

Specifically, we expect a nonlinear relationship between age and market value change (with younger players exhibiting larger positive changes), and a positive linear association between offensive performance metrics and market value growth.

# Important Metrics and Interpretation – Raw Datasets

This project begins with three raw datasets:

1. `players.csv` – Player demographic and market value snapshot  
2. `appearances.csv` – Match-level performance data  
3. `player_valuations.csv` – Historical market value time series  

Each dataset serves a distinct role and contains variables with different units, meanings, and limitations.

---

# Dataset 1: Players (players.csv)

This dataset contains one row per professional male soccer player, representing their most recent recorded season.

## Important Variables

### 1. market_value_in_eur  
**Unit:** Euros (nominal, not inflation-adjusted)

Represents the player's estimated current market value as reported by Transfermarkt.

Typical ranges:
- Youth/reserve players: €0–€1M  
- Mid-tier professionals: €1M–€20M  
- Top-league starters: €20M–€80M  
- Elite players: €80M–€200M+  

This is not a transaction price. It is a publicly estimated valuation and may reflect:
- Media exposure  
- Reputation  
- Recent performance  
- Contract length  

Values above €200M are rare and typically associated with global superstars.

---

### 2. highest_market_value_in_eur  
**Unit:** Euros  

The highest recorded estimated market value in the player's career history.  
This reflects peak perceived valuation and can serve as a proxy for career ceiling.

---

### 3. last_season  
**Unit:** Season end year (integer)

Represents the most recent season in which the player appeared in the dataset.  
For European leagues, season 2023 typically corresponds to the 2022–2023 season.

---

### 4. position  
Categorical variable describing player role (e.g., Forward, Defender, Goalkeeper, Midfielder).

Economic interpretation:
- Forwards typically command higher valuations due to scoring output.
- Goalkeepers often have lower volatility in valuation.
- Defensive players may be undervalued relative to attacking statistics.

---

### 5. date_of_birth  
Used to compute age.  
Age is critical because player value typically follows a lifecycle pattern:
- Rapid increase during early 20s
- Peak around 23–27
- Gradual decline after 28–30

---

### 6. height_in_cm  
**Unit:** Centimeters  

Physical attribute that may correlate with position (e.g., taller goalkeepers/defenders).

---

## Major Concerns (Players Dataset)

1. **Snapshot Bias**  
   This dataset captures players at their most recent season only. It does not provide full season-by-season market value history.

2. **Estimated Valuation, Not Real Prices**  
   Market value reflects perception rather than actual transfer fee.

3. **Missing Values**  
   Notable missingness exists in:
   - agent_name
   - contract_expiration_date
   - height_in_cm
   - market_value_in_eur (approx. 8–10%)

4. **Selection Bias**  
   Players tracked are mostly from major competitions. Lower-tier leagues may be underrepresented.

---

# Dataset 2: Appearances (appearances.csv)

This dataset contains match-level performance data. Each row represents one player appearance in one game.

## Important Variables

### 1. minutes_played  
**Unit:** Minutes  

Indicates how long the player participated in that match.  
Range:
- 0–15 → brief substitute appearance  
- 45 → half played  
- 90 → full match  
- 120 → includes extra time  

Total seasonal minutes reflect:
- Player importance  
- Injury status  
- Coach trust  

---

### 2. goals  
**Unit:** Count per match  

Number of goals scored in that match.  
For most players:
- 0 goals in majority of matches  
- Multi-goal games are rare  

Season totals:
- 0–3 → defenders  
- 5–15 → strong attacking contributors  
- 20+ → elite scoring season  

---

### 3. assists  
**Unit:** Count per match  

Number of credited assists in that match.  
Season totals:
- 0–3 → defensive players  
- 5–10 → productive creative players  
- 15+ → elite playmaking season  

---

### 4. yellow_cards / red_cards  
**Unit:** Count per match  

Disciplinary events.  
- Yellow cards indicate minor infractions  
- Red cards indicate serious misconduct or denial of scoring opportunity  

High card counts may signal aggressive style or disciplinary issues.

---

### 5. date  
Match date.  
Used to derive season (European season spans July–June).

---

## Major Concerns (Appearances Dataset)

1. **Match-Level Granularity**  
   Raw data is match-level. It must be aggregated to season-level before analysis.

2. **No Advanced Metrics**  
   The dataset does not include:
   - Expected goals (xG)
   - Expected assists (xA)
   - Key passes
   - Defensive actions  

   Without these, performance is measured only by realized outcomes (goals/assists), which may reflect finishing luck rather than underlying quality.

3. **Competition Heterogeneity**  
   Includes multiple competitions via `competition_id`. Differences in league quality may influence performance comparability.

4. **Survivorship Bias**  
   Only recorded appearances are included. Injured players or unused substitutes are underrepresented.

---

# Dataset 3: Player Valuations (player_valuations.csv)

This dataset contains historical market valuations over time.

## Important Variables

### 1. market_value_in_eur  
**Unit:** Euros (nominal)

Represents estimated market value at a specific date.

Because this dataset contains multiple valuation dates per player, it allows construction of:

- Season-end valuations  
- Season-to-season value changes  

---

### 2. date  
Valuation date.  
Used to assign valuation to a season (e.g., closest to June 30).

---

## Major Concerns (Valuations Dataset)

1. **Non-Transaction Based**  
   Market values are subjective estimates, not actual transfer prices.

2. **Irregular Valuation Timing**  
   Some players have many valuation updates; others have sparse updates.  
   This creates uneven temporal resolution.

3. **Market Hype Effects**  
   Media exposure, nationality, and club prestige may inflate perceived value independent of performance.

4. **Football-Specific Inflation**  
   Transfer markets inflate faster than general CPI.  
   Inflation adjustment using macroeconomic CPI may not fully capture football-specific price dynamics.

---

# Overall Structural Concerns

Across all raw datasets:

- The data reflects **professional male players in tracked competitions**, not the entire global soccer population.
- Lower leagues and less visible competitions may be underrepresented.
- Market value is a perceived economic measure, not a realized transaction price.
- Performance metrics are outcome-based (goals, assists) and lack advanced expected-value metrics.

---

# Summary

The raw datasets collectively provide:

- Player demographic and positional information
- Match-level performance data
- Historical market valuations

However, raw data requires:
- Season derivation from dates
- Aggregation to player-season level
- Selection of season-end valuations
- Inflation adjustment
- Handling of missing and irregular valuation updates

Only after these steps can meaningful season-to-season valuation dynamics be analyzed.


```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
import pandas as pd
import numpy as np
from pathlib import Path

RAW_DIR = Path("data/00-raw")
INTERIM_DIR = Path("data/01-interim")
PROCESSED_DIR = Path("data/02-processed")


players = pd.read_csv(RAW_DIR/"players.csv")
perf = pd.read_csv(RAW_DIR/"appearances.csv")

def zscore(s):
    s = pd.to_numeric(s, errors="coerce")
    return (s - s.mean()) / s.std(ddof=0)

def winsorize(s, p_low=0.01, p_high=0.99):
    s = pd.to_numeric(s, errors="coerce")
    lo, hi = s.quantile([p_low, p_high])
    return s.clip(lo, hi)
```

```python
p = players.copy()

# types
p["player_id"] = pd.to_numeric(p["player_id"], errors="coerce").astype("Int64")
p["last_season"] = pd.to_numeric(p["last_season"], errors="coerce").astype("Int64")
p["market_value_in_eur"] = pd.to_numeric(p["market_value_in_eur"], errors="coerce")
p["highest_market_value_in_eur"] = pd.to_numeric(p["highest_market_value_in_eur"], errors="coerce")
p["height_in_cm"] = pd.to_numeric(p["height_in_cm"], errors="coerce")

# dates
p["date_of_birth"] = pd.to_datetime(p["date_of_birth"], errors="coerce")
p["contract_expiration_date"] = pd.to_datetime(p["contract_expiration_date"], errors="coerce")

# position bucket
p["primary_position"] = p["position"]

# basic sanity flags
p["flag_missing_dob"] = p["date_of_birth"].isna()
p["flag_missing_pos"] = p["primary_position"].isna()
p["flag_bad_mv"] = (p["market_value_in_eur"] < 0) | (p["market_value_in_eur"] > 5e8)

# dedup in case of duplicates
p["_mv_isna"] = p["market_value_in_eur"].isna().astype(int)
p = (p.sort_values(["player_id", "last_season", "_mv_isna"])
       .drop_duplicates(subset=["player_id", "last_season"], keep="first")
       .drop(columns=["_mv_isna"]))

p.to_csv(INTERIM_DIR / "players_clean_interim.csv", index=False)
p.shape
```

```python
s = perf.copy()

# parse keys/types
s["player_id"] = pd.to_numeric(s["player_id"], errors="coerce").astype("Int64")
s["game_id"] = pd.to_numeric(s["game_id"], errors="coerce").astype("Int64")
s["minutes_played"] = pd.to_numeric(s["minutes_played"], errors="coerce")
s["goals"] = pd.to_numeric(s["goals"], errors="coerce")
s["assists"] = pd.to_numeric(s["assists"], errors="coerce")
s["yellow_cards"] = pd.to_numeric(s["yellow_cards"], errors="coerce")
s["red_cards"] = pd.to_numeric(s["red_cards"], errors="coerce")

# parse date
s["date"] = pd.to_datetime(s["date"], errors="coerce")

# drop rows without critical identifiers or date
s = s.dropna(subset=["player_id", "date"])

# derive season end year
s["season"] = np.where(s["date"].dt.month >= 7,
                       s["date"].dt.year + 1,
                       s["date"].dt.year).astype("int")

# keep only 2013–2023 seasons
s = s[s["season"].between(2013, 2023, inclusive="both")]

s.shape
```

```python
agg_cols = {
    "minutes_played": "sum",
    "goals": "sum",
    "assists": "sum",
    "yellow_cards": "sum",
    "red_cards": "sum",
    "game_id": "nunique",  # count matches appeared in
}

season_stats = (s.groupby(["player_id", "season"], as_index=False)
                  .agg(agg_cols)
                  .rename(columns={"game_id": "matches_played"}))

# per90 rates (only meaningful with enough minutes)
season_stats["minutes_played"] = season_stats["minutes_played"].clip(lower=0)

for c in ["goals", "assists", "yellow_cards", "red_cards"]:
    season_stats[f"{c}_per90"] = np.where(season_stats["minutes_played"] >= 450,
                                          90 * season_stats[c] / season_stats["minutes_played"],
                                          np.nan)

# z-score per season (your question says "standardized performance metrics")
for c in ["goals_per90", "assists_per90"]:
    season_stats[f"z_{c}"] = season_stats.groupby("season")[c].transform(
        lambda x: (x - x.mean()) / x.std(ddof=0)
    )

season_stats.to_csv(INTERIM_DIR / "season_stats_interim.csv", index=False)
season_stats.shape
```

```python
# Keep only needed player columns 
p_keep = p[["player_id", "date_of_birth", "primary_position"]].copy()

panel = season_stats.merge(p_keep, on="player_id", how="left")

# age at season end (June 30 of season end-year)
season_end = pd.to_datetime(panel["season"].astype(str) + "-06-30", errors="coerce")
panel["age"] = (season_end - panel["date_of_birth"]).dt.days / 365.25

# sanity flags check if ages are between 15 and 45
panel["flag_bad_age"] = (panel["age"] < 15) | (panel["age"] > 45)
panel["flag_missing_pos"] = panel["primary_position"].isna()

panel.to_csv(PROCESSED_DIR / "player_season_panel.csv", index=False)
panel.shape
```

```python
valu = pd.read_csv(RAW_DIR / "player_valuations.csv")

valu["player_id"] = pd.to_numeric(valu["player_id"], errors="coerce").astype("Int64")
valu["date"] = pd.to_datetime(valu["date"], errors="coerce")
valu["market_value_in_eur"] = pd.to_numeric(valu["market_value_in_eur"], errors="coerce")
valu = valu.dropna(subset=["player_id", "date", "market_value_in_eur"])

# derive season
valu["season"] = np.where(valu["date"].dt.month >= 7,
                          valu["date"].dt.year + 1,
                          valu["date"].dt.year).astype("int")
valu = valu[valu["season"].between(2013, 2023, inclusive="both")]

# pick one valuation per player-season: closest to June 30 (season end)
season_end_dates = pd.to_datetime(valu["season"].astype(str) + "-06-30", errors="coerce")
valu["days_from_season_end"] = (valu["date"] - season_end_dates).abs().dt.days

valu_season = (valu.sort_values(["player_id","season","days_from_season_end"])
                 .drop_duplicates(["player_id","season"], keep="first")
                 [["player_id","season","market_value_in_eur"]])

# inflation adjustment TODO: add accurate cpi data
cpi_map = {y: 1.0 for y in range(2013, 2024)}
cpi_2023 = 1.0

valu_season["cpi"] = valu_season["season"].map(cpi_map).astype(float)
valu_season["mv_real_2023eur"] = valu_season["market_value_in_eur"] * (cpi_2023 / valu_season["cpi"])

# merge valuations into panel
full = panel.merge(valu_season[["player_id","season","mv_real_2023eur"]], on=["player_id","season"], how="inner")

# compute consecutive-season changes
full = full.sort_values(["player_id","season"])
full["mv_prev"] = full.groupby("player_id")["mv_real_2023eur"].shift(1)
full["season_prev"] = full.groupby("player_id")["season"].shift(1)
full["is_consecutive"] = (full["season"] - full["season_prev"] == 1)

full["delta_log_mv"] = np.log1p(full["mv_real_2023eur"]) - np.log1p(full["mv_prev"])
full["delta_direction"] = (full["delta_log_mv"] >= 0).astype(int)

# keep modeling rows
model_df = full[full["is_consecutive"] & full["mv_prev"].notna()].copy()

# drop bad ages, missing positions, and low-minutes noisy seasons
model_df = model_df[~model_df["flag_bad_age"]]
model_df = model_df[~model_df["flag_missing_pos"]]
model_df = model_df[model_df["minutes_played"] >= 450]

# winsorize to reduce leverage
model_df["minutes_w"] = winsorize(model_df["minutes_played"], 0.01, 0.99)
model_df["delta_log_mv_w"] = winsorize(model_df["delta_log_mv"], 0.01, 0.99)

model_df.to_csv(PROCESSED_DIR / "mv_change_model_dataset.csv", index=False)
model_df.shape
```
