## Background and Prior Work


Professional soccer has become a global labor market where players are routinely assigned monetary values that influence transfers, wages, and long-term roster planning. While actual transfer fees are negotiated case-by-case (and only exist when a transfer occurs), many analysts use player “market value” estimates as a consistent proxy for how the market rates a player at a given moment. These estimates are widely used in sports analytics research because they are available for large numbers of players across many leagues and seasons, enabling statistical analysis of how age, performance, position, and context relate to value.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) 
A common public source for market values is Transfermarkt, which publishes crowd-informed estimates that explicitly should not be treated as the same thing as transfer fees. Transfermarkt describes its values as being informed by “pricing models” and, importantly, by structured community discussion moderated through a market value process—meaning the output reflects human judgment as well as available data.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)  This matters for our project because it implies that market value can capture real market signals (age, contract situation, reputation) but may also reflect biases in perception (e.g., hype, league visibility). In our analysis, we treat market value as an observed outcome variable and focus on identifying associations with measurable player attributes rather than claiming a single “true” value. 
Prior academic work has examined which factors explain player value and how accurately values can be estimated from performance and contextual indicators. For example, Müller et al. (2017) developed a data-driven approach to estimate market values for thousands of players across multiple top European leagues and quantified the effects of numerous indicators (including performance and context), highlighting that systematic modeling can reproduce meaningful value patterns and that both on-field output and situational factors matter.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) This line of work motivates our approach of combining player performance statistics with demographic and contextual variables (age, position, league/club context) using regression-style models to interpret which predictors relate most strongly to market value in our dataset. 
Researchers have also studied potential biases and structural effects that can show up in valuations. For instance, work on relative age effects (advantages linked to being born earlier in a selection year) found associations between relative age and market values in youth players, suggesting that development structures can influence who is perceived as valuable.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) This is relevant to our ethics and interpretation: even if performance metrics are included, market value may embed selection effects and scouting biases. Additionally, there are competing “algorithmic valuation” efforts (e.g., CIES/FIFA-discussed models) that incorporate factors like age and contract length, underscoring that valuation is multi-factor and contested—there isn’t one universally accepted formula.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) These studies frame our project’s goal: to empirically test which measurable attributes (performance, playing time, age, position, contract proxies when available) are most associated with market value, while acknowledging that values may reflect both performance and perception.
